## Segment 1 - Simple linear regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
rooms = 2 * np.random.rand(100,1) + 3
rooms[1:10]
```

```python
price = 256 + 6 * rooms + abs(np.random.rand(100,1))
price[1:10]
```

```python
plt.plot(rooms, price, 'r^')
plt.xlabel("num of rooms, 2019 ave")
plt.ylabel("2019 ave home, 1000s usd")
plt.show()
```

```python
X = rooms
y = price

LinReg = LinearRegression()
LinReg.fit(X,y)
print(LinReg.intercept_, LinReg.coef_)
```

*Simple Algebra*
- y = mx + b
- b = intercept = 265.7

*Estimated Coefficients*
- LinReg.coef_ = [5.99] Estimated coefficients for the terms in the linear regression problem.

```python
print(LinReg.score(X,y)) # r squred of our prediction
```
