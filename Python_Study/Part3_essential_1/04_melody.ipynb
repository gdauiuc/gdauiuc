## Segment 2 - Defining elements of a plot

```python
%matplotlib inline
rcParams['figure.figsize']= 5,4
```

### axes, ticks, and grids

```python
x = range(1,10)
y = [1,2,3,4,5,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5,6])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,6])

ax.grid()
ax.plot(x,y)
```

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)

ax1.plot(x)
ax2.plot(x,y)
```

## Segment 6 - Creating statistical data graphics

```python
import seaborn as sns
sb.set_style('whitegrid')
```

```python
address = 'Data/mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

cars.index = cars.car_names

mpg = cars['mpg']

mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sns.displot(data=mpg, kde=True) 
# since sns.distplot will be deprecated, future warning, so use displot and add parameters instead)
```

### scatter plot

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
sns.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

```python
sns.pairplot(cars)
```

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sns.pairplot(cars_subset)
plt.show()
```

### boxplots

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sns.boxplot(x='am', y='mpg', data=cars, palette='hls')
```
